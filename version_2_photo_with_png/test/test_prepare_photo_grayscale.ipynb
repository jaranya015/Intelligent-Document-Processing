```python
import os
import cv2
import matplotlib.pyplot as plt

# 1. กำหนดเส้นทางโฟลเดอร์ต้นทาง
input_folder = '/Users/faiijaran/Desktop/project/version_2/photo_for_test'

# 2. รายการนามสกุลไฟล์ที่รองรับ
valid_extensions = ('.jpg', '.jpeg', '.png', '.webp')

print(f"🚀 เริ่มประมวลผลและแสดงรูปภาพ (Preview Mode)...")

# ดึงรายชื่อไฟล์และเรียงลำดับ
files = [f for f in os.listdir(input_folder) if f.lower().endswith(valid_extensions)]
files.sort()

for filename in files:
    input_path = os.path.join(input_folder, filename)
    
    # อ่านรูปภาพ
    img = cv2.imread(input_path)
    
    if img is not None:
        # แปลงเป็น Grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # --- ส่วนแสดงผลใน Jupyter ---
        plt.figure(figsize=(10, 10))
        # ใช้ cmap='gray' เพื่อให้แสดงผลเป็นขาวดำถูกต้อง
        plt.imshow(gray_img, cmap='gray')
        plt.title(f"📄 File: {filename}")
        plt.axis("off") # ปิดตัวเลขแกน X, Y
        plt.show()
        
        print(f"✅ ประมวลผลเสร็จสิ้น: {filename}")
        print("-" * 50)
    else:
        print(f"❌ ไม่สามารถอ่านไฟล์: {filename}")

print(f"\n✨ แสดงผลครบทั้งหมด {len(files)} ไฟล์เรียบร้อยแล้ว!")
```

```python
import cv2
import matplotlib.pyplot as plt

def simple_grayscale(image_path):
    img = cv2.imread(image_path)
    if img is None: return None
    
    # แปลงเป็นขาวดำแบบไม่ใส่ฟิลเตอร์เบลอใดๆ ทั้งสิ้น
    # เพื่อรักษาความคมของขอบตัวอักษรในไฟล์ดิจิทัล
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    return gray

# ทดสอบโชว์ผลลัพธ์
test_file = '/Users/faiijaran/Desktop/project/version_2/photo_for_test/1.jpg'
result = simple_grayscale(test_file)

if result is not None:
    plt.figure(figsize=(12, 12))
    # ใช้ interpolation='none' เพื่อให้ Jupyter โชว์ความคมชัดที่แท้จริง ไม่ต้องเกลี่ยสี
    plt.imshow(result, cmap='gray', interpolation='none')
    plt.title("Sharp Grayscale (No Blur)")
    plt.axis("off")
    plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt

# 1. กำหนดเส้นทางโฟลเดอร์ต้นทาง
input_folder = '/Users/faiijaran/Desktop/project/version_2/photo_for_test'
valid_extensions = ('.jpg', '.jpeg', '.png', '.webp')

# 2. ตั้งค่า DPI ให้สูงขึ้นเพื่อให้เห็นรายละเอียดชัดเจน
plt.rcParams['figure.dpi'] = 150 

print(f"🚀 เริ่มประมวลผลและแสดงรูปภาพ (Sharp Preview Mode)...")

files = [f for f in os.listdir(input_folder) if f.lower().endswith(valid_extensions)]
files.sort()

for filename in files:
    input_path = os.path.join(input_folder, filename)
    img = cv2.imread(input_path)
    
    if img is not None:
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        # --- แสดงผลแบบไม่เกลี่ยสี (Interpolation None) ---
        plt.figure(figsize=(12, 12))
        # การใส่ interpolation='none' หรือ 'nearest' จะทำให้ตัวหนังสือคมเหมือนต้นฉบับ
        plt.imshow(gray_img, cmap='gray', interpolation='none')
        
        plt.title(f"📄 Sharp View: {filename}")
        plt.axis("off")
        plt.show()
        
        print(f"✅ ประมวลผลเสร็จสิ้น: {filename}")
        print("-" * 50)
    else:
        print(f"❌ ไม่สามารถอ่านไฟล์: {filename}")
```
