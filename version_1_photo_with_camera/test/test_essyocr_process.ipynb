```python
import os
import easyocr
import cv2

# 1. ตั้งค่า Path รูปภาพที่ผ่านการ Pre-process มาแล้ว
input_folder = '/Users/faiijaran/Desktop/project/output/photo_by_grayscale'

# 2. โหลดโมเดล
print("⌛ กำลังเตรียมโมเดล EasyOCR (Thai + English)...")
reader = easyocr.Reader(['th', 'en'])

def get_sorted_text(image_path):
    results = reader.readtext(image_path)
    
    # เรียงลำดับพิกัด Y (บนลงล่าง) และ X (ซ้ายไปขวา)
    results.sort(key=lambda x: (x[0][0][1], x[0][0][0]))
    
    lines = []
    current_line = ""
    last_y = -1
    threshold = 15 

    for (bbox, text, prob) in results:
        top_left_y = bbox[0][1]
        if last_y == -1 or abs(top_left_y - last_y) <= threshold:
            current_line += " " + text
        else:
            lines.append(current_line.strip())
            current_line = text
        last_y = top_left_y
        
    if current_line:
        lines.append(current_line.strip())
    return "\n".join(lines)

# 3. เริ่มดึงข้อมูลและแสดงผลที่คอนโซล
print(f"🚀 เริ่มประมวลผลรูปภาพจาก: {input_folder}")
print("=" * 50)

for i in range(1, 17):
    filename = f"grayscale_{i}.png"
    img_path = os.path.join(input_folder, filename)
    
    if os.path.exists(img_path):
        try:
            # ดึงข้อความแบบเรียงบรรทัด
            structured_text = get_sorted_text(img_path)
            
            # --- แสดงผลลัพธ์ทั้งหมดออกที่หน้าจอ ---
            print(f"📍 [ไฟล์: {filename}]")
            print("-" * 20)
            if structured_text.strip():
                print(structured_text)
            else:
                print("(ไม่พบข้อความในรูปนี้)")
            print("-" * 50)
            
        except Exception as e:
            print(f"❌ พลาดที่ไฟล์ {filename}: {e}")
            print("-" * 50)
    else:
        print(f"⚠️ ไม่พบไฟล์: {filename}")

print("\n✨ ตรวจสอบข้อมูลด้านบนได้เลยครับ!")
```
